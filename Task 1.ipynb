### Problem Statement:
                          In this task we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. 
                          This is a simple Linear regression task as it involves just two variables (Hours & Score).
                            

## Importing the required Libraries

```python
import matplotlib.pyplot as plt  #used for plotting
import numpy as np # used for working with arrays
import pandas as pd  #used for data analysis

# SKlearn : Simple and efficient tools for predictive data analysis
from sklearn.linear_model import LinearRegression   
from sklearn.model_selection import train_test_split
from sklearn import metrics  
```

```python
url = 'http://bit.ly/w-data'
data = pd.read_csv(url)
```

## Analyzing Data

```python
# checking first 5 rows of the dataset
data.head()
```

```python
# getting the shape of dataset
data.shape
```

```python
# checking for NULL values
data.isnull().sum()
```

```python
data.describe()
```

### Ploting the DataSet 

```python
# plotting a scatterplot to represent the correlation
data.plot(x='Hours', y='Scores', style='o', figsize=(9,7))  
plt.title('Study Hours vs Percentage Obtained')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

#### The above graph shows a clear positive linear correlation between Hours Studied and Percentage Scores

## Preparation of Data

```python
# Prepairing the data for distribution into the test and train dataset 
# to develop our model.
X = data.iloc[:, :-1].values  
Y = data.iloc[:, 1].values

# checking for X and Y dims
X.shape, Y.shape
```

## Splitting DataSet into train & test sets

```python
# since the dataset is less, let X and Y be splitted in 8:2 ratio
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.20, random_state=0)
```

## Building the Linear Regression Model

```python
# initialisation of model from LinearRegression
model = LinearRegression()
# training the regression model with the train dataset
model.fit(X_train,Y_train)
```

```python
# Plotting a scatterplot for X and Y (Training Set)
line=model.coef_*X+model.intercept_
plt.title('Train Set Regression Line')
plt.xlabel('No. Hours')
plt.ylabel('Scores')
plt.scatter(X,Y)
plt.plot(X,line)
plt.show()
```

## Model Prediction

```python
# validating predictions
y_pred=model.predict(X_test) 
```

```python
# Comparing Actual vs Predicted Scores
compare_data = pd.DataFrame({'Actual': Y_test, 'Predicted': y_pred})  
compare_data
```

### Bar Plot Visulaization

```python
compare_data.plot(kind='bar', figsize=(9,6))
plt.title('Visual Comparision of Actual and Predicted Values of Scores')
```

## Predicting the Score

```python
hours = 9.25
pred = model.predict([[hours]])
print("The Predicted score of the Student is:", pred)
```

## Evaluation of Performence Metrics of Model

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test,y_pred))
print('Mean Squares Error:', metrics.mean_squared_error(Y_test,y_pred))
print('R2 Accuracy:', metrics.r2_score(y_true=Y_test,y_pred=y_pred)*100,)
```

## Conclusion :
### The Model is perfectly Trained and it's accuracy is 94.54% which is quite good
